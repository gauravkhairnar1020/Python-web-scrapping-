```python
soup = BeautifulSoup(page.text)
soup
```

```python
table = soup.find('table')
```

```python
table = table.find('tbody')
```

```python
rows = table.find_all('tr',style='')
len(rows)
```

```python
column_info =[]
for i in rows:
    cols=i.find_all('td')
#     print(cols)
    crypto_info = [c.text for c in cols]
#     print([c.text for c in cols])
    column_info.append(crypto_info)
```

```python
for i in column_info:
    if (i[1].isdigit()):
        print(i[1],i[2],i[3],i[4],i[5],i[6],i[7],i[8],i[9])
```

*For each element, "i", in the iterable, the code checks if the second element in "i" (i[1]) is a digit using the string method "isdigit()". If it is a digit, the code then prints the second through tenth elements of "i" (i[1] through i[9]).


```python
Name = []
Price = []
OneHour = []
TwentyFourHours = []
SevenDays = []
Market_Cap = []
Volume_TwentyFourHours = []
Circulating_Supply = []
for i in column_info:
    if i[1].isdigit():
        Name.append(i[2])
        Price.append(i[3])
        OneHour.append(i[4])
        TwentyFourHours.append(i[5])
        SevenDays.append(i[6])
        Market_Cap.append(i[7])
        Volume_TwentyFourHours.append(i[8])
        Circulating_Supply.append(i[9])
```

This script defines eight empty lists,
Then it iterates through an iterable called "column_info", for each element "i" in the iterable, the script checks if the second element in "i" (i[1]) is a digit using the string method "isdigit()",if it is, the script then appends the second through the ninth element of "i" to the respective lists..
This is a way to separate the data from column_info list into 8 different lists based on the second element of the element being digit.





```python
Name
```

```python
data = {
    'Name': Name, 
    'Price': Price, 
    '1h %': OneHour, 
    '24h %': TwentyFourHours,
    '7d %': SevenDays,
    'Market Cap': Market_Cap,
    "Volume(24h)": Volume_TwentyFourHours,
    'Circulating Supply': Circulating_Supply,
}
```

*This dictionary combines all the lists into one data structure, which allows for easy and organized access to the data. With this data structure, we can easily access and use the data in a structured way.





```python
import pandas as pd
```

```python
Crypto_market = pd.DataFrame(data)
```
